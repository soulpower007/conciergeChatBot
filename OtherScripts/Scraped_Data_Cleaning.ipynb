```python
# f.close()

import json
f = open('data_clean.json', 'r')

data = json.load(f)

mp = {}
count=0
for i in data:
    print(i["id"])
    count+=1
    mp[i["id"]]=i

f.close()


print(len(mp), count)
```

```python
ret = []
for i in mp.values():
#     print(i)
    ret.append(i)
```

```python
with open('data_preprocessed.json', 'w') as f:
    json.dump(ret,f)
```
